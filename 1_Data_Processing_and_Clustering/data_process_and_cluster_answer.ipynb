### **🩺 关键医学指标详解**

**1. 胸痛类型 (cp)**：
- 0：典型心绞痛（运动诱发，休息缓解）
- 1：非典型心绞痛（症状不典型）
- 2：非心绞痛（其他原因引起的胸痛）
- 3：无症状（无明显胸痛）

**2. 静息心电图 (restecg)**：
- 0：正常心电图
- 1：ST-T波有异常（心肌缺血表现）
- 2：根据Estes准则，有潜在的左心室肥厚

**3. ST段变化特征**：
- **oldpeak**：运动导致ST段下移的程度，数值越大表示心肌缺血越严重
- **slope**：ST波峰值的斜率
  - 0：上升（相对少见）
  - 1：平坦（常见于心肌缺血）
  - 2：下降（心肌缺血表现）

**4. 地中海贫血症 (thal)**：
- normal：正常
- fixed defect：固定缺陷（永久性心肌损伤）
- reversable：可逆缺陷（暂时性心肌缺血）

**常见处理方法**：

# **第二部分：数据标准化（为聚类做准备）**

## **2.1 为什么需要数据标准化？**

**重要概念**：数据标准化是机器学习中非常关键的一步！

**医学含义**：
- 不同检查项目的单位不同：血压单位是mmHg（数值约100-150），胆固醇单位是mg/dl（数值约200-300）
- 如果不标准化，大数值的特征（胆固醇）会"主导"分析，而小数值的特征（血压）会被忽视
- 这就像医生给不同检查指标赋予相同的"重要性权重"

**标准化方法**：将所有特征转换为相同的尺度（通常是均值0，标准差1）

````python
# 先看看各特征的数值范围
print("📊 各特征数值范围：")
print(df_cleaned.describe())

# 可视化不同特征的分布（直方图）
numeric_features = ['age', 'trestbps', 'chol', 'thalach']
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('不同医学指标的数值分布情况', fontsize=16)

# 为直方图设置颜色
hist_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']
feature_titles = ['年龄 (岁)', '静息血压 (mmHg)', '胆固醇 (mg/dl)', '最大心率 (bpm)']

for i, feature in enumerate(numeric_features):
    row, col = i // 2, i % 2
    
    # 绘制直方图
    axes[row, col].hist(df_cleaned[feature], bins=20, color=hist_colors[i], 
                      edgecolor='black', alpha=0.7)
    
    # 添加均值线
    mean_value = df_cleaned[feature].mean()
    axes[row, col].axvline(mean_value, color='red', linestyle='--', 
                          linewidth=2, label=f'均值: {mean_value:.1f}')
    
    axes[row, col].set_title(feature_titles[i], fontsize=12, fontweight='bold')
    axes[row, col].set_xlabel('数值', fontsize=10)
    axes[row, col].set_ylabel('频数', fontsize=10)
    axes[row, col].legend(fontsize=9)
    axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("📈 直方图分析：")
print("- 直方图显示每个特征的数值分布情况")
print("- 红色虚线表示该特征的均值")
print("- 观察不同特征的分布形状和范围差异")
````

## **2.2 执行数据标准化**

我们将使用**Z-Score标准化**（也称为Standardization）：

**公式**：
- 新值 = (原值 - 均值) / 标准差
- 标准化后：均值≈0，标准差≈1

**医学类比**：将不同检查项目的数值都转换为"标准分数"，这样医生可以公平地比较不同指标。

````python
# 选择需要标准化的数值型特征（只选择数值列）
# 排除target列（这是标签，不需要标准化）
numeric_columns = df_cleaned.select_dtypes(include=[np.number]).columns.tolist()
numeric_columns.remove('target') if 'target' in numeric_columns else None

features_to_scale = df_cleaned[numeric_columns]

# TODO: 创建标准化器并执行标准化
# 提示：
# 1. 创建StandardScaler对象
# 2. 使用fit_transform方法对数据进行标准化
scaler = StandardScaler()
df_scaled = scaler.fit_transform(features_to_scale)

# 将标准化后的数据转换为DataFrame（为了方便显示）
df_scaled = pd.DataFrame(df_scaled, columns=numeric_columns)

print("✅ 标准化完成！")
print(f"\n📊 使用的数值特征：{numeric_columns}")
print("\n📊 标准化后的数据示例（前5行）：")
print(df_scaled.head())
````

# **第三部分：K-Means聚类算法**

## **3.1 什么是聚类？**

**基本概念**：
- **聚类**是无监督学习方法，不需要预先知道标签
- 目标：将相似的患者分到同一组
- 每组称为一个"簇"（cluster）

**医学应用场景**：
- 将心脏病患者按严重程度分组
- 发现新的疾病亚型
- 个性化治疗方案制定
- 流行病学研究中的人群分层

## **3.2 K-Means算法原理**

**算法名称含义**：
- **K**：预设的聚类数量
- **Means**：使用均值（平均数）来表示聚类中心

**算法步骤（医学类比）**：

**第1步：随机选择K个初始中心**
- 类比：在患者群体中随机选K个人作为"代表患者"

**第2步：分配患者到最近的中心**
- 计算每个患者与K个中心的"距离"
- 将患者分配到最近的中心所在组
- 类比：医生将患者分配给最擅长治疗其病症的专家

**第3步：更新中心位置**
- 计算每个组中所有患者的平均特征值
- 将平均值作为新的中心位置
- 类比：根据新患者的病情，专家更新自己对"典型病例"的认识

**第4步：重复步骤2-3，直到中心不再变化**
- 类比：经过多轮调整，每个专家的"典型患者"特征趋于稳定

## **3.3 确定聚类数量K**

**如何选择K值?**

**医学类比**:
- K=2:患者分为"高风险"和"低风险"两组
- K=3:分为"高风险"、"中风险"、"低风险"三组
- K值选择会影响分组的临床意义

**常用方法**:
- **领域知识**:根据医学先验知识(如已知有3种心脏病类型)
- **肘部法则**:寻找效果提升放缓的转折点
- **轮廓系数**:衡量聚类质量的指标

````python
# 使用肘部法则确定最佳K值
inertia_values = []  # 存储每个K值的SSE(误差平方和)
k_range = range(1, 11)  # 测试K从1到10

print("🔍 正在测试不同的K值...")
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia_values.append(kmeans.inertia_)
    print(f"K={k}: SSE={kmeans.inertia_:.2f}")

# 绘制肘部图
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia_values, 'bo-', markersize=8)
plt.xlabel('聚类数量 K', fontsize=12)
plt.ylabel('误差平方和 (SSE)', fontsize=12)
plt.title('肘部法则:确定最佳K值', fontsize=14)
plt.grid(True, alpha=0.3)

# 标记肘部点
plt.axvline(x=3, color='r', linestyle='--', alpha=0.7, label='可能的肘部点 (K=3)')
plt.legend()

plt.show()

print("\n🤔 观察肘部图:")
print("- K=3附近曲线斜率明显放缓,这可能是合理的聚类数量")
print("- 我们选择K=3进行后续分析(对应低、中、高风险三组)")
````

## **3.4 执行K-Means聚类**

现在我们使用设置的K值对患者进行聚类分析。

````python
K = 3 # 可自行设置，也可以设置为2，3，5 ……

# TODO: 创建并训练K-Means模型
# 提示:
# 1. 创建KMeans对象,使用上面设置的K值
# 2. 使用fit方法对标准化后的数据进行聚类
kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
kmeans.fit(df_scaled)

# 获取聚类结果
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

print("✅ K-Means聚类完成!")
print(f"\n📊 聚类结果统计:")
print(f"- 聚类数量:{kmeans.n_clusters}")
print(f"- 迭代次数:{kmeans.n_iter_}")
print(f"- 最终误差:{kmeans.inertia_:.2f}")

# 统计每个簇的患者数量
cluster_counts = pd.Series(cluster_labels).value_counts().sort_index()
print("\n👥 各组患者数量:")
for cluster_id, count in cluster_counts.items():
    percentage = (count / len(cluster_labels)) * 100
    print(f"- 簇 {cluster_id}:{count} 人 ({percentage:.1f}%)")
````
